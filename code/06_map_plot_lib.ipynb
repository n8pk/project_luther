```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# map graph

```python
houses = pd.read_pickle('/home/nate/ds/metis/class_work/projects/project_luther/data/den_with_cord.pkl')
distros = pd.read_pickle('/home/nate/ds/metis/class_work/projects/project_luther/data/den_distro_cord.pkl')
```

```python
house_cords = list(houses['cords'])
distro_cords = distros
```

```python
h_lat = np.array([float(x[0]) for x in house_cords])
h_lon = np.array([float(x[1]) for x in house_cords])

d_lat = np.array([float(x[0]) for x in distro_cords])
d_lon = np.array([float(x[1]) for x in distro_cords])
```

```python
plt.figure(figsize=(20,20))
plt.plot(h_lat, h_lon, 'bo', d_lat, d_lon, 'gs')
plt.title('Houses (Blue) and Dispensaries (Green)', fontsize=24, fontweight='bold')
plt.ylabel('Longitude', fontsize=14, fontweight='bold')
plt.xlabel('Latitude', fontsize=14, fontweight='bold')
plt.show();
            
            
```
